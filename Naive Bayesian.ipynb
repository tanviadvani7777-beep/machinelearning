```python
import pandas as pd
from sklearn.naive_bayes import CategoricalNB
from sklearn.preprocessing import LabelEncoder
```

```python
data = {
    'Age'    : ['Youth', 'Youth', 'Middle', 'Senior', 'Senior', 'Senior', 'Middle', 'Youth', 'Youth', 'Senior', 'Youth', 'Middle', 'Middle', 'Senior'],
    'Income' : ['High', 'High', 'High', 'Medium', 'Low', 'Low', 'Low', 'Medium', 'Low', 'Medium', 'Medium', 'Medium', 'High', 'Medium'],
    'Student': ['No', 'No', 'No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No'],
    'CreditRating': ['Fair', 'Excellent', 'Fair', 'Fair', 'Fair', 'Excellent', 'Excellent', 'Fair', 'Fair', 'Fair', 'Excellent', 'Excellent', 'Fair', 'Excellent'],
    'BuysComputer': ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'Yes', 'Yes', 'Yes', 'No']
}
```

```python
df=pd.DataFrame(data)
print(df)
```

```python
#1. Encode categorical numerical
le = {}  
df_enc = df.copy() 

for col in df.columns:  
    le[col] = LabelEncoder() 
    df_enc[col] = le[col].fit_transform(df[col])  

df_enc
```

```python
#2. feature & target
x = df_enc[['Age','Income','Student','CreditRating']]
y = df_enc['BuysComputer']
```

```python
#3. model training
model = CategoricalNB()  
model.fit(x, y)  

```

```python
# test data
test = [['Youth', 'Medium', 'Yes', 'Fair']]
test
```

```python
# 4. Prediction 
print("\n~ Prediction ---------------------------~")
for t in test:
    row = [[
        le['Age'].transform([t[0]])[0],
        le['Income'].transform([t[1]])[0],
        le['Student'].transform([t[2]])[0],
        le['CreditRating'].transform([t[3]])[0]
    ]]
    
    pred = le['BuysComputer'].inverse_transform(model.predict(row))[0]
    prob = model.predict_proba(row)[0]
    
    print(f"Input : Age={t[0]}, Income={t[1]}, Student={t[2]}, CreditRating={t[3]}")
    print(f"Result: Buys Computer -> {pred}")
    print(f"print('P(No)  = {prob[0]:.4f}')")
    print(f"print('P(Yes) = {prob[1]:.4f}')")
```
