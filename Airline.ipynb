```python
import pandas as pd
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import CategoricalNB

df = pd.read_csv('Airline Dataset Updated - v2.csv')
df
```

```python
def categorize_age(age):
    if age <= 30:
        return 'Youth'
    elif age <= 60:
        return 'Middle'
    else:
        return 'Senior'
df['Age_Group'] = df['Age'].apply(categorize_age)
df_subset = df[['Age_Group', 'Gender', 'Continents', 'Flight Status']].copy()
```

```python
#1. Encode categorical numerical
le = {}
df_enc = df_subset.copy()

for col in df_subset.columns:
    le[col] = LabelEncoder()
    df_enc[col] = le[col].fit_transform(df_subset[col])
```

```python
#2. feature & target
X = df_enc[['Age_Group', 'Gender', 'Continents']]
y = df_enc['Flight Status']
```

```python
#3. model training
model = CategoricalNB()
model.fit(X, y)
```

```python
# 4. Prediction 
test = [['Youth', 'Female', 'Europe']]

print("\n~ Prediction ---------------------------~")
for t in test:
    # Encode the test sample inputs using the fitted label encoders
    row = [[
        le['Age_Group'].transform([t[0]])[0],
        le['Gender'].transform([t[1]])[0],
        le['Continents'].transform([t[2]])[0]
    ]]
    
    # Predict the class outcome and inverse transform back to text
    pred_encoded = model.predict(row)[0]
    pred = le['Flight Status'].inverse_transform([pred_encoded])[0]
    
    # Calculate probabilities for each class
    prob = model.predict_proba(row)[0]
    
    print(f"Input : Age_Group={t[0]}, Gender={t[1]}, Continents={t[2]}")
    print(f"Result: Flight Status -> {pred}\n")
    
    print("Class Probabilities:")
    for idx, class_name in enumerate(le['Flight Status'].classes_):
        print(f"  P({class_name:<9}) = {prob[idx]:.4f}")
```
